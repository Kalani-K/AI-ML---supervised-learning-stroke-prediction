```python
#Normalize only currently relevant numeric Features
scaler = MinMaxScaler()
df[["Age", "Average Glucose Level", "Body Mass Index (BMI)", "Stress Levels"]] = scaler.fit_transform(
    df[["Age", "Average Glucose Level", "Body Mass Index (BMI)", "Stress Levels"]]
)

df[["Age", "Average Glucose Level", "Body Mass Index (BMI)", "Stress Levels"]].describe()
```

```python

plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Average Glucose Level"])
plt.title("Boxplot of Average Glucose Level")
plt.xlabel("Average Glucose Level")
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix of Numeric Features")
plt.tight_layout()
plt.show()
```
