```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
#loading the dataset
df = pd.read_csv('../AI-ML Group Project/Dataset/stroke_prediction_dataset.csv')
#".." means to go up one directory
```

```python
#View the top 5 rows of the dataset
display(df.head())
```

```python
#Checking the dataframe's dimensions, data types & non missing value counts
df.info()
```

```python
#Only showing the columns with missing values
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0] #Filter columns with missing values > 0
print(missing_values)
```

```python
#Handling missing values for "Symptoms" column
if df["Symptoms"].isnull().sum() > 0:
    #Adding a binary indicator as a missing indicator
    df["Symptoms_missing"] = df["Symptoms"].isnull().astype(int)

    #Filling in missing values as "Uknown"
    df.fillna({"Symptoms":"Unknown"}, inplace = True)
```

```python
#Dataframe after handling missing values:
df.isnull().sum()
```

```python
#Visualizing top 12 rows after change
df.head(12)
```

```python
#As showm above only the "Symptoms" column had missing values, 2500 rows to be precise

#A binary indicator column "Symptoms_missing" was added as in many cases, missingness itself can carry predictive power
#for instance, patients who are too sick to report symptoms might have worse outcomes.

#Affected rows were filled with "Unknown" as someone may not have reported symptoms due to death, incapacity, or incomplete records.
```
